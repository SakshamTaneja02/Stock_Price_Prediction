```python
import yfinance as yf #this package enables us to work with yahoo finance API
```

```python
sp500 = yf.Ticker("^GSPC") #This is ticker class that enables us to download stock prices and indices for a single
# symbol (in our case GSPC which S&P500)
```

```python
sp500 = sp500.history(period="max") # query the historical prices for period max ie when index was created
```

```python
sp500 # we get a pandas data frame
```

```python
sp500.plot(y="Close",use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500
```

```python
# so we will use ML to predict whether the price will go up or down tomorrow
# we will do this by training the model on the days the price goes up and then predicting it for the day we want
sp500["Tomorrow"] = sp500["Close"].shift(-1) #so closing value for next day will become the tomorrow value for today
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy() #we will take the data for only recent times as the stock market may have 
# through some changes in the time period and old data may not be useful in out prediction
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier # we will use random forest as they are resistant to overfitting
#they tend to run quickly and can pick up non-linear tendencies in data eg open price is not linearly related to target
```

```python
model = RandomForestClassifier(n_estimators=500,min_samples_split=100,random_state=1)
#n_estimators gives us the number of individual decision tree we want the higher the better 
# min_samples_split specifies the minimum number of samples required to split an internal node, also responsible 
# for preventing overfitting the higher the less accurate our model will be but less it will overfit

train = sp500[:-100]
test = sp500[-100:]

predictors = ["Close","Volume","Open","High","Low"] 

model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score
```

```python
preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds,index=test.index)
preds
```

```python
precision_score(test["Target"],preds)
```

```python
combined = pd.concat([test["Target"],preds],axis=1)
combined
```

```python
combined.plot()
```

```python
# we are now going to build a more robust way to test our algo against multiple years so we are going to do 
# backtesting
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds,index=test.index,name="Predictions")
    combined = pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
# so for eg our start is 10yrs and step is 1yr so this means that we will take data for first 10yrs and predict for
# 11thyr, then we will take first 11yrs of data and predict for 12yr and so on
# this way we will get predictions for a lot of years
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    for i in range(start, data.shape[0],step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500,model,predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
horizons = [2,5,60,250,1000] # we will calculate the mean in last two days, last five days and so on and find ratio
# bewtween todays closing price and closing price in these periods which will help us look how much the market has 
# gone up or gone down
new_predictors = [] # to test our model on some new predictors

for horizon in horizons:
    rolling_avg = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] /  rolling_avg["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column,trend_column]
```

```python
sp500
```

```python
sp500 = sp500.dropna()
sp500
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds = model.predict_proba(test[predictors])[:,1] # we are now taking the probability that it will go up or down
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds,index=test.index,name="Predictions")
    combined = pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
predictions = backtest(sp500,model,new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```
